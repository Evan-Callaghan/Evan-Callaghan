```python
import pandas as pd
import numpy as np
import io

## Importing data set
from google.colab import files
uploaded = files.upload()
```

```python
blockbuster = pd.read_csv(io.BytesIO(uploaded['blockbuster.csv']))

blockbuster.head()
```

```python
blockbuster[blockbuster['Title'] == 'Breaking Bad'][['Title', 'Languages', 'Genre', 'Summary', 'Popularity_Score']]
```

```python
blockbuster[blockbuster['Title'] == 'Breaking Bad'][['Title', 'Languages', 'Genre', 'Summary', 'Popularity_Score']]
```

```python
blockbuster['Series or Movie'].value_counts()
```

```python
blockbuster['Languages'].value_counts()
```

```python
## Creating a smaller subset of the blockbuster data set

spanish = blockbuster[blockbuster['Languages'] == 'Spanish'][['Title', 'Genre',	'Series or Movie',	'View Rating', 	'Summary', 'Popularity_Score']]
```

```python
spanish['Series or Movie'].value_counts()
```

```python
import altair as alt
alt.Chart(spanish).mark_bar().encode(
  x='count(Series or Movie)',
  y='Series or Movie',
  color='Series or Movie'
)
```

```python
import altair as alt
alt.Chart(spanish).mark_bar().encode(
  x=alt.X('Popularity_Score', bin=True),
  y='count()',
)
```

```python
english = blockbuster[blockbuster['Languages'] == 'English'][['Title', 'Genre',	'Series or Movie',	'View Rating', 	'Summary', 'Popularity_Score']]
```

```python
import altair as alt
alt.Chart(english).mark_bar().encode(
  x='count(Series or Movie)',
  y='Series or Movie',
  color='Series or Movie'
)
```
